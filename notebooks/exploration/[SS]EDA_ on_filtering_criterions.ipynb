## What if we can find bad annotators by those that given divergent views from other annotators frequently?

## Lets see if we can find annotators who often disagree with the other two?

```python
df_uniquelabels = df.groupby("pair_id")["reduced_label"].nunique()
pairs_twoagree = list(df_uniquelabels[(df.groupby("pair_id")["reduced_label"].nunique() == 2).values].index)
df_twoagree = df[df["pair_id"].isin(pairs_twoagree)]
```

#### Random Sample #1

```python
df[df['annotator']=="A1NRZ7O1TEJB1V"]
```

This persons rating overall seem to be very shoddy - lets see if we can isolate people who do it often

```python
df_id_reducedlabel = df_twoagree.groupby("pair_id")['reduced_label'].median()
df_twoagree['generally_accepted_label']  = df_id_reducedlabel.values.repeat(3)
df_twoagree[df_twoagree.reduced_label != df_twoagree.generally_accepted_label].groupby('annotator').size()
```

```python
df_twoagree[df_twoagree["pair_id"] == "pair_9503"]
```

```python
df_twoagree[df_twoagree['annotator']=='A110SC5K5Y3IHS']
```

```python
df_unpopularopinion = df_twoagree[df_twoagree.reduced_label != df_twoagree.generally_accepted_label].groupby('annotator').size().reset_index()
df_unpopularopinion.columns = ['annotator','unpopular_opinion']

df_allopinions = df[df['annotator'].isin(list(df_unpopularopinion.annotator))].groupby('annotator').size().reset_index()
df_allopinions.columns = ['annotator','all_opinion']

df_opinion_all_unpop = df_allopinions.merge(df_unpopularopinion,on="annotator")
```

I randomly assigned half of the opinions to unpopular for it to be a sign...

```python
df_opinion_all_unpop[((df_opinion_all_unpop.unpopular_opinion / df_opinion_all_unpop.all_opinion) > 0.5) & (df_opinion_all_unpop.all_opinion > 4)]
```

```python
df[df['annotator'] == 'AFQIHS5XX58NR']
```

## This seems to be a good direction - it isn't a perfect method but seems to help guide us that while we don't have a clear idea what everyone is basing their opinions off, some people clearly skew from popular opinion.

```python
df_uncool_kids = df_opinion_all_unpop[((df_opinion_all_unpop.unpopular_opinion / df_opinion_all_unpop.all_opinion) > 0.5) & (df_opinion_all_unpop.all_opinion > 4)]['annotator']
```

```python
sns.distplot(df.groupby('annotator')['total_seconds'].mean(),label="avg")
sns.distplot(df[df['annotator'].isin(list(df_uncool_kids))].groupby('annotator')['total_seconds'].mean(),label="uncool")
plt.legend();
plt.show();
```

## What is we shifted gears and looked at people to see if they can fit into "Radical" and "Centrist" categories (inclined to 1,5 or 2,4 labeling)

```python
df_classify_annotator = df.copy()
```

```python
min_five_posters = list(pd.Series(df.groupby('annotator').size()).mask(lambda x: x < 5).dropna().index)
df_classify_annotator = df_classify_annotator[df_classify_annotator['annotator'].isin(min_five_posters)]
```

We should remove the 3's because they dont tell us about peoples preferences

```python
df_classify_annotator_remove3 = df_classify_annotator[df_classify_annotator['label'] != 3]
```

```python
sns.distplot(df_classify_annotator_remove3.groupby('annotator')['label'].var().values);
```

```python
df_classify_annotator_remove3['radical'] = [1 if ((x ==1) or (x==5)) else -1 if ((x==2) or (x==4)) else 0 for x in df_classify_annotator_remove3.label]
```

```python
sns.distplot(df_classify_annotator_remove3.groupby('annotator')['radical'].mean());
```

It does seem to suggest that it is more common for there to be radical than otherwise.

There are however, a few points of contention:
1.   Often the radical labels are due to "bad actor" labeling (as in, people who just throw a 1,5 without proper examination)
2.   There are annotators who are only exposed to random datasets, and therefore will only score a 1 because there is a bias to randomness.
     (This can be avoided is we only look at radicality between random datasets and non-random datasets)

```python
df_radical_random = df_classify_annotator_remove3.groupby('random')
for name, group in df_radical_random:
    sns.distplot(group.groupby('annotator')['radical'].mean(), label=name);
plt.legend();
plt.show()
```

There does seem to be good justification here to smooth out peoples scores, as they are either "radically radical" or "radically centrist"

Would this still hold when we remove bad actors and bad labels?

# Summary



1.   Length of the sentences doesn't seem to have any relevance to time of answer or label score.
2.   Looking at annotators based off their average (or median) rating speed doesn't give us any indication either, even when we control for outliers.
3.   Looking at individual ratings duration (those done extremely fast/slow) might be an avenue to explore more deeply - but unlikely.

We might be able to use these general features as a way to "rate the raters" - an indication of how long they take, taken in conjuction with other features, might guide us further...

4.   By reducing the score to: -1 dissimilar, 0 unsure, 1 similar:
*   We can look at pairs where all three disagree and explore why this might be.
*   We can find patterns of people who are often give an unpopular opinion (disagree when both other annotators agree)

The fourth point seems to be the best direction, and it could be a space to explore:

*   Filtering out "uncertain" pairs or "bad actor" annotators
*   Look at those cases to help understand where ambiguity within text pairs.
*   What is the overlap between people "unpopular opinion" annotators and other meta-features (like avg. rating duration)
*   Explore ignoring label = 3, as it is equivacle to -1 in other cases (unknown vs error)

5.   By filtering out random datasets (which seems to have a high bias to labels=1 - understandably), we see that looking at radical labeling, there is strong evidence that there is a large enough population of people who are either "radically radical" or "radically centrist", that label smoothing might be beneficial.
